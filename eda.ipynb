```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data into a pandas dataframe
df = pd.read_excel('lukoil-data.xlsx')

# Show the first few rows of the data
print(df.head())

# Display basic statistics of the numerical variables
print(df.describe())

# Visualize the distribution of each variable using histograms
df.hist(bins=10, figsize=(20,15))
plt.show()

# Visualize the pairwise relationships between numerical variables using scatter plots
sns.pairplot(df, diag_kind='hist')
plt.show()

# Visualize the correlation between numerical variables using a heatmap
sns.heatmap(df.corr(), cmap='coolwarm', annot=True)
plt.show()

# Visualize the distribution of each variable by origin using box plots
for col in df.columns[:-1]:
    sns.boxplot(x=col, y='origin', data=df)
    plt.show()

# Visualize the pairwise relationships between numerical variables by origin using scatter plots
sns.pairplot(df, hue='origin', diag_kind='hist')
plt.show()

# Visualize the correlation between numerical variables by origin using a heatmap
for o in df['origin'].unique():
    sns.heatmap(df[df['origin']==o].corr(), cmap='coolwarm', annot=True)
    plt.title(o)
    plt.show()
```

```python
# read the data into a pandas dataframe
data = pd.read_excel('REDO-DATA.xlsx')

# iterate over each variable and plot a Gaussian bell curve
for column in data.columns:
    plt.hist(data[column], density=True, bins=30)

    # fit a normal distribution to the data
    mu, sigma = np.mean(data[column]), np.std(data[column])
    x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
    plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) *
             np.exp(- (x - mu)**2 / (2 * sigma**2)), linewidth=2, color='r')
    
    plt.title(column)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
```
